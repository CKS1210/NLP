## Encoder-Decoder
The Encoder-Decoder architecture is a framework used in sequence-to-sequence (seq2seq) models for various tasks involving sequential data, such as machine translation, text summarization, and more. It consists of two main components: an encoder and a decoder.

#### Workflow in Sequence-to-Sequence Tasks:
* **Encoding:** The input sequence is fed into the encoder, which processes and encodes it into a `fixed-size context vector`.
* **Context Vector:** The context vector encapsulates the input sequence's information and serves as the initial state for the decoder.
* **Decoding:** The decoder takes the context vector and generates the output sequence by predicting tokens one step at a time, `considering the context vector and previously generated tokens`.
